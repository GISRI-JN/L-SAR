```python
import pandas as pd
import numpy as np
from datetime import datetime
 
# Excel
df = pd.read_excel('D:\weather_Data\.xlsx', sheet_name='tp')
#
df['time'] = df['time'].dt.year.astype(str)
 
# DataFrame
data_array = df.values
#
#data_array[:,0] = pd.to_datetime(data_array[:, 0])
#print(data_array)
#data_array[:,0] = data_array[:,0].dt.strftime("%Y")
#print(data_array)
 
lats = np.arange(36,38.1,0.1)
lats = np.around(lats, decimals=1)

lons = np.arange(116,118.1,0.1)
lons = np.around(lons, decimals=1)

years = np.arange(2014,2024,1)

arr_result = np.empty((1,5))
for lat in lats:
    lat_array = data_array[ data_array[:,1] == lat]
    for lon in lons:
        lon_array = lat_array[ lat_array[:,2] == lon]
        value_array = lon_array[:,3]
        if(value_array.size == 0):
            continue

        #
        #min = np.amin(value_array)
        #max = np.amax(value_array)
        #average  = np.mean(value_array)

        #10
        year_rain = np.empty(10, dtype=float)
        #
        for year in years:
            #print(year)
            #print(lon_array[:,0] )
            year_array  = lon_array[   lon_array[:,0] == str(year) ]
            
            #print(year_array)
            
            rains = year_array[:,3]
            #print(rains)
            total = np.sum(rains)
            year_rain[year-2014] = total

        print(year_rain)

        #
        min_year = np.amin(year_rain)
        max_year = np.amax(year_rain)
        average_year  = np.mean(year_rain)
        #print(min_year)
        #print(max_year)
        #print(average_year)
        
        #print(lat,lon,min,max,np.around(max-min,decimals=3))
        data = np.array([lat,lon,min_year,max_year,average_year])
        arr_result = np.vstack((arr_result,data))

print(arr_result.shape)
print(arr_result.size)
print(arr_result)

csv_name = 'D:\\weather_Data\\.csv'
import csv
 
# CSV
with open(csv_name, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    for row in arr_result:
        writer.writerow(row)
 
```
